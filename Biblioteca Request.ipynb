# Biblioteca Request

Serão apresentados o uso dos Métodos **GET**, **POST**, **PATCH** e **DELETE**

***API Utilizada:*** *https://docs.awesomeapi.com.br/api-de-moedas*  
***Base Utilizada:*** "estudorequisicao" *do Firebase (https://console.firebase.google.com/u/0/?pli=1)*

### Médoto GET - Pegando Informações

```python
import requests

#Requisicao da API
#requisicao = requests.get('https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL')

#requisicao do FireBase para o banco estudorequisicao, quando utiliza ele, precisa colocar no final o .json
requisicao = requests.get('https://estudorequisicao-default-rtdb.firebaseio.com/.json')

print(requisicao) #Response [200] é que deu certo e não gerou erro
print(requisicao.json())
```

### Médoto POST - Criando Informações

```python
#para utilizar o método POST é necessario informar qual infromação quer criar
#para isso, utilizaremos a variável dados e precisa ser em json
dados = '{"Nome": "Malibu"}'

#coloca a variável dados no final da linha alimentando o "data"
requisicao = requests.post('https://estudorequisicao-default-rtdb.firebaseio.com/.json',data=dados)

print(requisicao)
print(requisicao.json())
```

### Médoto PATCH - Editando Informações

```python
#para utilizar o método PATCH é necessario informar qual infromação quer editar
#para isso, utilizaremos endereco da pessoa que quero atualizar que foi criado no POST
dados = '{"Nome": "Malibu", "Sobrenome": "Aleixo Amorim"}'

#coloca a variável dados no final da linha alimentando o "data"
requisicao = requests.patch('https://estudorequisicao-default-rtdb.firebaseio.com/-Msacg0cyRmoENibBoqc.json',data=dados)

print(requisicao)
print(requisicao.json())
```

### Médoto DELETE - Deletando Informações

```python
#requisicao do FireBase para o banco estudorequisicao, quando utiliza ele, precisa colocar no final o .json
requisicao = requests.delete('https://estudorequisicao-default-rtdb.firebaseio.com/-Msacg0cyRmoENibBoqc.json',data=dados)

print(requisicao)
print(requisicao.json())
```
